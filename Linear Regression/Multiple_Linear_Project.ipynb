```python
## Create the dataframe
dataset = pd.DataFrame(california.data, columns=california.feature_names)
dataset.head()
```

```python
dataset['Price'] = california.target
```

```python
## Indepdent and Dependent Features
X = dataset.iloc[:,:-1] ## Indepedent Features
Y = dataset.iloc[:,-1] ## Dependent Feature
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.33,random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
Y_pred_test = model.predict(X_test)
```

```python
Y_pred_test
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
mse = mean_squared_error(Y_test,Y_pred_test)
mae = mean_absolute_error(Y_test,Y_pred_test)
score = r2_score(Y_test,Y_pred_test)
print("MSE : ", mse)
print("MAE : ", mae)
print("r2 Score : ", score)
```

```python
## Assumptions
plt.scatter(Y_test, Y_pred_test)
plt.xlabel("True test Data")
plt.ylabel("Predicted Y Data")
```

```python
residuals = Y_test - Y_pred_test
plt.scatter(Y_pred_test, residuals)
```

```python
sns.distplot(residuals, kde=True)
```

## pickling

- Python Pickle Module is used for serializing and deserializing a python object structure, any object can be pickled so that it can be saved on the disk. what pickle does is that it serializes the object before wrting it to a file. Pickling is a way to convert a python object into a character stream. the idea is that this character stream contains all the information necessary to reconstruct the object into another python script.

```python
import pickle
pickle.dump(model, open('regressor.pkl','wb'))
```

```python
model = pickle.load(open('regressor.pkl','rb'))
```

```python
model.predict(X_test)
```
