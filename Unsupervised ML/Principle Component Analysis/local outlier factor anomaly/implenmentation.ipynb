```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.neighbors import LocalOutlierFactor
```

```python
from sklearn.datasets import make_blobs

x, _ = make_blobs(n_samples=300,centers=1,cluster_std=0.60,random_state=42)
## add some outliers 
X_outliers = np.random.uniform(low=-6, high=6, size=(20, 2))
X_combined = np.vstack([x, X_outliers])
```

```python
lof = LocalOutlierFactor(n_neighbors=5, contamination=0.05)  # contamination â‰ˆ expected proportion of outliers
y_pred = lof.fit_predict(X_combined)  # -1 for outliers, 1 for inliers
scores = lof.negative_outlier_factor_
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_combined[:, 0], X_combined[:, 1], c=y_pred, cmap='coolwarm', edgecolors='k')
plt.title("Local Outlier Factor (LOF) Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```

```python
outlier_indices = np.argsort(scores)[:10]  # top 10 outliers
print("Top outlier indices:", outlier_indices)
```
