```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
iris['DESCR']
```

```python
X = pd.DataFrame(iris.data,columns=['sepal_length','sepal_width','petal_length','petal_width'])
```

```python
X
```

```python
Y = iris.target
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train , Y_test = train_test_split(X,Y,test_size=0.20,random_state=42)
```

```python
### apply descision tree classifier
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=2)
clf.fit(X_train,Y_train)
```

```python
Y_pred = clf.predict(X_test)
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
print(classification_report(Y_test,Y_pred))
print(confusion_matrix(Y_test, Y_pred))
print(accuracy_score(Y_test, Y_pred))
```

```python
## Visualize the decision tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(clf, filled=True)
```

```python
## overfitting has been done with this classification technique so apply post or pre prunning
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
print(classification_report(Y_test,y_pred))
print(confusion_matrix(Y_test, y_pred))
print(accuracy_score(Y_test, y_pred))
```

## Decesion Tree Classifier Preprunning and Hyperparameter tuning 

```python
param = {
    'criterion' : ['ArithmeticError', 'gini', 'entropy'],
    'splitter' : ['best','random'],
    'max_depth' : [1,2,3,4,5],
    'max_features' : ['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
treemodel = DecisionTreeClassifier()
```

```python
grid = GridSearchCV(treemodel,param_grid=param,cv=5,scoring='accuracy')
```

```python
grid.fit(X_train,Y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
print(classification_report(Y_test,grid.predict(X_test)))
```

```python
print(confusion_matrix(Y_test,grid.predict(X_test)))
```
