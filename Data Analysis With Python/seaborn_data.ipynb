```python
import seaborn as sns

```

```python
# Basic Plotting with Seaborn
tips = sns.load_dataset('tips')
tips
```

```python
## create a scatter plot
import matplotlib.pyplot as plt
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips)
plt.title('Scatter Plot of Total Bill VS Tips')
plt.show()
```

```python
## line plot 
sns.lineplot(x = 'size', y = 'total_bill', data = tips)

plt.show()
```

```python
# # matplotlib and seaborn can also be used for categorical ploting of datasets.
### Categorical Plots 
## Bar plot

sns.barplot(x = 'day', y = 'total_bill', data = tips)
plt.title('This is Bar Plot of Day VS Total Bill')
```

```python
## Box Plot

sns.boxplot(x = 'day', y = 'total_bill', data = tips)
plt.title('Box Plot of Day VS Total Bill')
```

```python
## Violin Plot 
sns.violinplot(x = 'day', y = 'total_bill', data = tips)
plt.title('Violin Plot of Size VS Total Bill')
plt.show()
```

```python
## histograms 
sns.histplot(tips['total_bill'], bins = 10, kde = True)

## with the help of kde i can get the line for distirbution of data.
```

```python
## KDE Plot
sns.kdeplot(tips['total_bill'], fill = True)
```

```python
## Pair Plot
sns.pairplot(tips)
## we specifically need to have numerical data in attributes for this particular pairplot data visualization.
```

```python
## Heat Map
corr = tips[['total_bill', 'size', 'tip']].corr()
corr
sns.heatmap(corr,annot = True, cmap='coolwarm')
```
