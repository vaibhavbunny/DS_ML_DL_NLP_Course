```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [10,20,30,40,50,60]

plt.plot( x, y)
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.title('Graph')
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.plot(x,y,color = 'red', linestyle = '--',marker = 'o', linewidth = 1,markersize = 9)
plt.grid(True)
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
z = [1,8,27,64,125]

plt.figure(figsize = (9,5))
plt.subplot(2,2,1)
plt.plot(x,y,color = 'Blue')
plt.title('Plot 1')

plt.subplot(2,2,2)
plt.plot(x,z,color = 'Red')
plt.title('Plot 2')

plt.subplot(2,2,3)
plt.plot(y,x,color = 'Orange')
plt.title('Plot 3')

plt.subplot(2,2,4)
plt.plot(z,x,color = 'Red')
plt.title('Plot 4')
```

```python
## Bar Plot
categories = ['A','B','C','D','E']
values = [1,2,3,4,5]
plt.bar(categories,values,color = 'Red')
```

```python
## histograms are used to depict the distribution in data
## data is divided into bins and count the number of data points in each bin

x = [1,1,2,3,3,3,3,3,3,4,4,4,5,5,6,6,6,6,6,6,6,6,6,6,6,6,6,6,6,6,7,7,8,9,9,9,9]

plt.hist(x,bins=5,color='Red',edgecolor = 'Black')
```

```python
## create a scatter plot
x = [1,2,3,4,5]
y = [2,3,4,5,6]

plt.scatter(x,y,color = 'Red',marker= 'x')
```

```python
## pie chart 

data = [10,20,40,20,30]
lables = ['A','B','C','D','E']
color = ['Red','Green','Blue','Yellow','lightskyblue']
explode =[0.2,0,0,0,0]
plt.pie(data,labels=lables,colors=color,autopct="%1.1f%%",explode=explode,shadow=True)
```
